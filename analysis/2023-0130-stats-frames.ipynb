```python
STD_THH = 4.5
vis_df_copy_all = pd.concat([
    pd.read_parquet(f'./../stat_analysis_result/vis_df_copy_std{STD_THH}_head{headi}.parquet').assign(head=headi)
#     pd.read_parquet(f'./../stat_analysis_result/vis_df_copy_{STD_THH}_{headi}.parquet')
    for headi in range(4)
])
    
    
```

```python
vis_df_copy_all['p_notCDR_TCR_Bond']
```

```python
import numpy as np
from scipy import stats

result_table = []
for headi in range(4):
    vis_df_copy = vis_df_copy_all.query('head==@headi')
    for c in vis_df_copy.columns:
        if c in ['count_bonds_to_notself_TCRchain', 'head', 'p_opposite', 'p_anyHbond', 'p_hbond_to_notself_TCRchain']:
            continue
        rrr = {}
        rrr['property'] = c
        vals = vis_df_copy[c]
        mean_large_attention = np.nanmean([float(v[0]) for v in vals if v[0] is not None])
        mean_small_attention = np.nanmean([float(v[1]) for v in vals if v[0] is not None])
        rrr['large'] = mean_large_attention
        rrr['small'] = mean_small_attention
#         print(f'mean_large_attention {mean_large_attention:.4f}, mean_small_attention {mean_small_attention:.4f}')
#         print(stats.ttest_rel([float(v[0]) for v in vals if v[0] is not None],
#                                  [float(v[1]) for v in vals if v[0] is not None]))
        pv = stats.ttest_rel([float(v[0]) for v in vals if v[0] is not None],
                                 [float(v[1]) for v in vals if v[0] is not None], 
                             nan_policy='omit'
                            ).pvalue
        rrr['pvalue'] = pv
        
        if pv < 0.05:
            rrr['star'] = '**'
        else:
            rrr['star'] = ''
            
        rrr['head'] = headi
        result_table.append(rrr)

```
