```python
def query_json(pep):
    return """{"query": {
    "type": "group",
    "logical_operator": "and",
    "nodes": [
      {
        "type": "terminal",
        "service": "full_text",
        "parameters": {
          "value": "{pep}"
        }
      }
    ]
  },
  "return_type": "assembly"
}""".replace('{pep}',pep).replace('\n','')
```

```python

def query_json_motif(motif):
    return """{
  "query": {
    "type": "terminal",
    "service": "seqmotif",
    "parameters": {
      "value": "PEPTIDEPLACEHOLDER",
      "pattern_type": "simple",
      "target": "pdb_protein_sequence"
    }
  },
  "return_type": "polymer_entity"
}""".replace('PEPTIDEPLACEHOLDER', motif).replace('\n','')


PEPSEQ = 'GILGFVFTL'

query = f'''https://search.rcsb.org/rcsbsearch/v1/query?json={query_json_motif(PEPSEQ)}'''
print('query=', query)

import requests
response = requests.get(query)
print(response.json())
```

```python
PEPSEQ = 'GILGFVFTL'

def query_json_motif(motif):
    return """{
  "query": {
    "type": "terminal",
    "service": "seqmotif",
    "parameters": {
      "value": "PEPTIDEPLACEHOLDER",
      "pattern_type": "simple",
      "target": "pdb_protein_sequence"
    }
  },
  "return_type": "polymer_entity"
}""".replace('PEPTIDEPLACEHOLDER', motif).replace('\n','')



query = f'''https://search.rcsb.org/rcsbsearch/v1/query?json={query_json_motif(PEPSEQ)}'''

def get_idlist_from_peptide(pepseq):
    query = f'''https://search.rcsb.org/rcsbsearch/v1/query?json={query_json_motif(pepseq)}'''
    response = requests.get(query)
    return [v['identifier'].split('_')[0] for v in response.json()['result_set']]
```
